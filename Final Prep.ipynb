## Demographics

```python
dem_df = pd.read_csv('../data/demographics.csv')
dem_df.info()
```

```python
dem_df.drop(columns=['Country','State'], inplace=True)
dem_df.head()
```

```python
#lets see how education affects churn

(
    dem_df.groupby('Education')
    .agg({'CustomerNo':'count'})
)
```

```python
#How does gender affect churn
(
    dem_df.groupby('Gender')
    .agg({'CustomerNo':'count'})
)
```

```python
Ser_DF = pd.read_csv('../data/services.csv')
Ser_DF.info()
```
