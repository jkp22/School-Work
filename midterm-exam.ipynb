## Question 1
Write a program using the "f-strings" (https://realpython.com/python-f-strings/) and user input function to calculate Body mass index (BMI). 

Formula: $BMI=\frac {weight (kg)} {[height (m)]^2}$

Hint: you may need int() or float() function

An example output:
```
Weight (kg): 68
Height (cm): 165
BMI: 24.977043158861342
```

````python
# Complete your code below (4 points) 
w = input('Weight (kg): ')
h = input('Height (cm): ')
h = int(h)
bmi = int(w) / h**2
bmi = bmi * 10000
print(f'Your BMI is, {bmi}!')
````

### Step 2. Data Analysis

````python
# 3.5
# How many unique brands are there in this dataset? (1 point)
df['Brand'].nunique()
````

````python
# 3.6
# Which destination (country) has the most hotels listed in this dataset? (1 points)
# List the number of hotels in each country. 
(
    df.groupby('Destination')
    .agg({'Hotel':'count'})
    .sort_values('Hotel',ascending=False)
    
)
````

````python
# 3.7
# How many hotels in China have their category changes? (2 points)
(
    df.loc[df.Destination=='China']
    .groupby('Hotel')
    .agg({'New Category':'max'})
)
# 68
````

````python
# 3.8
# How many brands in China have their category changed? (2 points)
(
    df.loc[df.Destination=='China']
    .groupby('Brand')
    .agg({'New Category':'max'})
)
#14
````

````python
# 3.9
# How many hotels worldwide were upgraded vs. downgraded in category in 2020? (2 points)
# Upgrade: "New Category" is greater than "Current Category".


df['Upgrade'] = df['New Category'] > df['Current Category']
(
    df.loc[df.Upgrade== True]
    .groupby(['Hotel','Upgrade'])
    .count()
    .filter(['Hotel','Upgrade'])
)
````

````python
# 3.10
# List all JW Marriott hotels in China that had a category upgrade. (2 points)
(
    df.loc[(df.Brand=='JW Marriott')&(df.Destination=='China')&(df.Upgrade==True)]
   
)
````

````python
df
````

````python
# 3.11
# List hotels whose category changes greater than 1 if any, (2 points)
# such as changing from category 3 to 5 or from category 7 to 4.
df['Grade_Dif'] = df['New Category'] - df['Current Category']
(
    df.loc[(df.Grade_Dif!=1)&(df.Grade_Dif!=0)&(df.Grade_Dif!=-1)]
)
````

````python
# 3.12
# What's the standard price for each category?  (2 points)
# Sort the eight categories in ascending order of price. 
# Hint: you can use either new or old category. 
(
    df.groupby(['Current Category'])
    .agg({'Current Standard Price':'max'})
    .reset_index()
    .sort_values('Current Standard Price', ascending=True)
)
````

````python
# 3.13
# Which brands had the most hotels upgraded in category? (3 points)
# List the top 5 brands and the corresponding number of hotels upgraded.
(
    df.loc[df.Upgrade==True]
    .groupby(['Brand'])
    .agg({'Grade_Dif':'count'})
    .reset_index()
    .sort_values('Grade_Dif', ascending=False)
    .head(5)
)
````

````python
# 3.14
# Group all hotels by destination and brand,
# Display the number of hotels, average current category, and average new category
# Only keep the destinations and brands with over 10 hotels. (3 points)
(
    df.groupby(['Brand','Destination'])
    .agg({'Hotel':'count','Current Category':'mean','New Category':'mean'})
    .reset_index()
    .query('Hotel>=10')
)
````
